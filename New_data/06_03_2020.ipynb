```python
dupes_2 = []
dupes_3 = []
dupes_4 = []

for j in range(train_1.__len__()):
    
    first_item = np.array(train_1.loc[j])
    
    #print("Item num ", j)

    for i in range(train_2.__len__()):
        sec_item = np.array(train_2.loc[i])

        if np.allclose(first_item, sec_item):
            #print("Second ", i)
            dupes_2.append(i)

    for i in range(train_3.__len__()):
        thir_item = np.array(train_3.loc[i])

        if np.allclose(first_item, thir_item):
            #print("Third ", i)
            dupes_3.append(i)

    for i in range(train_4.__len__()):
        fort_item = np.array(train_4.loc[i])

        if np.allclose(first_item, fort_item):
            #print("Forth ", i)
            dupes_4.append(i)
```
